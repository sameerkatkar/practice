```python
in_array = [0, math.pi/2, math.pi/3, np.pi]
print("Radian values : \n", in_array)

out_array = np.rad2deg(in_array)
print("Degree values: \n", out_array)
```

```python
degree = np.arange(10.) *90
print("Degree values: \n", degree)
radian = np.deg2rad(degree)
print("Radian values: \n", radian)
```

```python
data = {'C':20,'C++':15,'Java':30,'Python':50}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize=(10,5))
plt.bar(courses, values, color = 'green')
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
